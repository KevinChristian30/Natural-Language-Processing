```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Womens Clothing E-Commerce Reviews.csv')
df.columns
```

```python
columns_to_drop = ['Clothing ID', 
                   'Age', 
                   'Title', 
                   'Recommended IND', 
                   'Positive Feedback Count', 
                   'Division Name', 
                   'Department Name', 
                   'Class Name',
                   'Unnamed: 0'
                ]

df = df.drop(columns_to_drop, axis=1)
df
```

```python
df = df.dropna()
df
```

```python
positive = df[df['Rating'] > 2.5]
negative = df[df['Rating'] < 2.5]

positive.drop(['Rating'], axis=1)
negative.drop(['Rating'], axis=1)

positive
```

```python
positive.to_csv('positive.csv', index=False)
negative.to_csv('negative.csv', index=False)
```
